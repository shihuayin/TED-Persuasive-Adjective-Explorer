## 2.1 Defining Scraping and Extraction Functions

The first step is to define a function to extract the video links, titles, and description information from the CSV file containing the URLs of the "Most Popular TED Talks" for the past 10 years provided by TED.

The following code simulates the behavior of a browser, using the requests library to send HTTP requests to get the page content, and then uses the BeautifulSoup library to parse the HTML structure and extract the relevant 'a' tags to get the video links, titles, and descriptions. Finally, all the extracted data is consolidated into a Pandas DataFrame and saved as a new CSV file named url_ten_years_all_videos.csv (you can find it in the data set I submitted) for further data analysis or other processing.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

file_path = 'url_ten_years.csv'  
urls_df = pd.read_csv(file_path)

# Request headers to simulate a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'TE': 'Trailers'
}

# Store all scraped data
all_talks = []

# Perform scraping for each URL
for url in urls_df['URL']:
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        talks = []
        for a_tag in soup.find_all('a', class_='group block', href=True):
            link = "https://www.ted.com" + a_tag['href']
            
           # Find the title
            title_tag = a_tag.find('h4', class_='text-lg xl:text-2xl font-bold tracking-normal transition-colors group-hover:text-gray-700')
            title = title_tag.text.strip() if title_tag else 'No title found'
            
            # Find the description
            description_tag = a_tag.find('div', class_='text-sm')
            description = description_tag.text.strip() if description_tag else 'No description found'
            
            talks.append({'link': link, 'title': title, 'description': description})
        
        all_talks.extend(talks)
    else:
        # Log the failure to retrieve the page
        print(f"Failed to retrieve the page: {url}. Status code: {response.status_code}")

talks_df = pd.DataFrame(all_talks)

output_path = 'url_ten_years_all_videos.csv'  
talks_df.to_csv(output_path, index=False)
print(f"Data has been saved to {output_path}")
```

## 2.2 Scraping Related Topics for 100 TED Talk Links

The second step is to visit each video page from the 'url_ten_years_all_videos.csv' file and extract the related topics information from the page.

The following code uses the requests library to simulate browser behavior by sending HTTP requests to get the page content. After that, it uses the BeautifulSoup library to parse the HTML structure, locate the section containing the related topics, and consolidate them into a new DataFrame column. Finally, the updated data is saved as a new CSV file named all_videos_with_topics.csv (you can find it in the dataset I submitted).


```python
file_path = 'url_ten_years_all_videos.csv' 
urls_df = pd.read_csv(file_path)

# Request headers to simulate a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'TE': 'Trailers'
}

# Define a list to store related topics
related_topics_list = []

# Perform scraping for each URL
for url in urls_df['link']:
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        related_topics = []
        related_topics_section = soup.find('ul', class_='mb-6 inline-block text-black')
        if related_topics_section:
            for topic_tag in related_topics_section.find_all('a', class_='inline-block py-1 text-tui-sm capitalize underline'):
                related_topics.append(topic_tag.text.strip())
        
        # Add related topics to the list
        related_topics_list.append(', '.join(related_topics))
    else:
        # Log the failure to retrieve the page
        print(f"Failed to retrieve the page: {url}. Status code: {response.status_code}")
        related_topics_list.append('')

# Append the related topics to the DataFrame
urls_df['Related Topics'] = related_topics_list

output_path = 'all_videos_with_topics.csv' 
urls_df.to_csv(output_path, index=False)
print(f"Data has been saved to {output_path}")
```

## 2.3 Extracting Year Data

The following code extracts the year information from each TED Talks video link and adds it as a new column ('Year') to the DataFrame. The reason for extracting the year data is that I want to conduct time-based analysis in the subsequent analysis.


```python
import re

file_path = 'all_videos_with_topics.csv'
df = pd.read_csv(file_path)

# Function to extract year from link
def extract_year(link):
    match = re.search(r'(\d{4})', link)
    return match.group(0) if match else None

# Apply the function to the 'link' column and create a new 'Year' column
df['Year'] = df['link'].apply(extract_year)

# Save the updated DataFrame to a new CSV file
output_file_path = 'all_videos_with_topics_and_year.csv'
df.to_csv(output_file_path, index=False)

print("year OK")
print(f"Data has been saved to {output_file_path}")
```

## 2.4 Extracting Subtitles for Each Video

The third step is to modify the link format for each video from 'all_videos_with_topics_and_year.csv', add subtitle parameters to form new links, and scrape the subtitle text data.

The following code uses Selenium and Chrome WebDriver. The subtitles of each video are combined into a single string and stored in a new column called 'subtitles'. Intermediate results are saved every 10 videos to prevent data loss due to network interruptions or long operations. This strategy ensures that part of the processed data is retained even during long-running processes.

Finally, all data is saved as 'ted_talks_with_subtitles.csv' (you can find it in the dataset I submitted).


```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

csv_file_path = 'all_videos_with_topics_and_year.csv'
df = pd.read_csv(csv_file_path)

# Set up WebDriver
service = Service(executable_path="/Users/2shihua1yin/Downloads/chromedriver-mac-x64/chromedriver")
chrome_options = Options()
chrome_options.add_argument('--headless') 
driver = webdriver.Chrome(service=service, options=chrome_options)

# Add a new column to store subtitles
df['subtitles'] = ''

count = 0

# Iterate over all URLs in the CSV file
for index, row in df.iterrows():
    url = row['link'].replace("?referrer=", "/transcript?referrer=") + "&subtitle=en"
    driver.get(url)

    # Wait for the page to load
    time.sleep(5)  

    subtitles = driver.find_elements(By.CSS_SELECTOR, 'div[data-testid="paragraphs-container"] div[role="button"] span[dir="ltr"]')

    subtitle_text = ' '.join([subtitle.text.strip() for subtitle in subtitles])
    df.at[index, 'subtitles'] = subtitle_text

    # Print progress
    count += 1
    print(f"Processed {count}/{len(df)}: {url}")

    # Save every 10 scraped subtitles
    if count % 10 == 0:
        temp_csv_file_path = f'ted_talks_with_subtitles_part_{count//10}.csv'
        df.iloc[:count].to_csv(temp_csv_file_path, index=False)
        print(f"Intermediate CSV file saved to: {temp_csv_file_path}")

# Close the browser
driver.quit()

final_csv_file_path = 'ted_talks_with_subtitles.csv'
df.to_csv(final_csv_file_path, index=False)

print(f"Final CSV file saved to: {final_csv_file_path}")
```

### 3.4 Removing Stop Words

The following code further cleans and optimizes the data by removing stop words from the subtitles. The purpose of this is to improve data quality and provide more meaningful data for subsequent natural language processing and text analysis.


```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
# Get English stop words
stop_words = set(stopwords.words('english'))

# Define a function to remove stop words
def remove_stopwords(words):
    new_words = [word for word in words if word.lower() not in stop_words]
    return new_words

dataframe['subtitles_without_stopwords'] = dataframe['tokenized_subtitles'].apply(remove_stopwords)

# Display the first few rows of the processed data to confirm stop words have been removed correctly
print("First few rows of the data after removing stop words:")
print(dataframe[['tokenized_subtitles', 'subtitles_without_stopwords']].head(3))
```

### 3.5 Lemmatization

The following code performs lemmatization on the video subtitles. The specific steps include defining functions to get part-of-speech tags and lemmatize the words, applying lemmatization to the list of words after removing stop words, and storing the results in a new column called 'lemmatized_subtitles'.


```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()

# Define a function to get WordNet part-of-speech tags
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Define a function for lemmatization
def lemmatize_words(words):
    new_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
    return new_words


dataframe['lemmatized_subtitles'] = dataframe['subtitles_without_stopwords'].apply(lemmatize_words)

# Display the first few rows of the processed data to confirm lemmatization has been done correctly
print("First few rows of the data after lemmatization:")
print(dataframe[['subtitles_without_stopwords', 'lemmatized_subtitles']].head(4))
```

**Lemmatization reduces word forms to their base or root form, improving the consistency and analyzability of the data. You can see from the few rows printed that "runs" is lemmatized to "run," and "used" is lemmatized to "use."**

The following code classifies the subtitles of TED Talks into 'technology', 'science', 'business', or 'other' based on the 'Related Topics' column by defining a classification function. Then, it merges the word lists in the 'lemmatized_subtitles' column for each category to generate a new DataFrame named merged_df.


```python
# Create a new column to store classification information
def classify_topic(topics):
    if 'technology' in topics.lower():
        return 'technology'
    elif 'science' in topics.lower():
        return 'science'
    elif 'business' in topics.lower():
        return 'business'
    else:
        return 'other'

new_dataframe['category'] = new_dataframe['Related Topics'].apply(classify_topic)

# Filter the data for the three categories
technology_df = new_dataframe[new_dataframe['category'] == 'technology']
science_df = new_dataframe[new_dataframe['category'] == 'science']
business_df = new_dataframe[new_dataframe['category'] == 'business']

# Merge the lemmatized_subtitles column
technology_lemmas = technology_df['lemmatized_subtitles'].sum()
science_lemmas = science_df['lemmatized_subtitles'].sum()
business_lemmas = business_df['lemmatized_subtitles'].sum()

# Create a new DataFrame to store the merged data
merged_df = pd.DataFrame({
    'category': ['technology', 'science', 'business'],
    'lemmatized_subtitles': [technology_lemmas, science_lemmas, business_lemmas]
})

# Display the processed data to confirm the merging was done correctly
print("Merged data:")
print(merged_df)
```

```python

import spacy 
nlp = spacy.load("en_core_web_sm")

def extract_adjectives(text):
    doc = nlp(text)
    adjectives = [token.text for token in doc if token.pos_ == 'ADJ']
    return adjectives

# Extract adjectives from each category
adjectives_data = []
for index, row in merged_df.iterrows():
    adjectives = extract_adjectives(row['lemmatized_subtitles'])
    adjectives_data.append({'category': row['category'], 'adjectives': adjectives})


# Extract adjectives from each category
technology_adjectives = extract_adjectives(technology_lemmas)
science_adjectives = extract_adjectives(science_lemmas)
business_adjectives = extract_adjectives(business_lemmas)

# Create a new DataFrame to store the adjectives data
adjectives_df = pd.DataFrame(adjectives_data)

# Display the adjectives data to confirm extraction was done correctly
print("Extracted adjectives:")
print(adjectives_df)
```

The following code extracts unique adjectives for each category ('technology', 'science', 'business') from the TED Talks subtitle data:

### 4.4 Sentiment Analysis

The following code uses the VADER sentiment analysis tool to perform sentiment analysis on the subtitle data of each video, extracting negative, neutral, positive, and compound sentiment scores.
The analysis results are then stored in a new CSV file named final.csv (you can find it in the dataset I submitted) for use in the fifth section of the analysis.


```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

# Creating an instance of the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Defining the sentiment analysis function
def sentiment_analysis(text):
    sentiment_scores = analyzer.polarity_scores(text)
    return sentiment_scores['neg'], sentiment_scores['neu'], sentiment_scores['pos'], sentiment_scores['compound']

# Performing sentiment analysis on subtitles of each talk and storing the results in new columns
new_dataframe[['neg_sentiment', 'neu_sentiment', 'pos_sentiment', 'compound_sentiment']] = new_dataframe['subtitles'].apply(lambda x: pd.Series(sentiment_analysis(x)))


new_dataframe.to_csv('final.csv', index=False, encoding='utf-8')
print("Data saved as final.csv")
```


### 5.4 Visualizing Mean Sentiment Scores by Category

The following code calculates and visualizes the mean sentiment scores for different categories, showing the negative, neutral, positive, and compound sentiment scores for each category:
1. Group by category and calculate the mean of these sentiment scores.
2. Plot a heatmap to display the results.


```python
# Create a new DataFrame excluding the 'other' category
filtered_other_df = final_df[final_df['category'] != 'other']

# Group by category and calculate the mean sentiment scores
sentiment_means = filtered_other_df.groupby('category').agg({
    'neg_sentiment': 'mean',
    'neu_sentiment': 'mean',
    'pos_sentiment': 'mean',
    'compound_sentiment': 'mean'
}).reset_index()

# Display the results
print(sentiment_means)

# Prepare data for the heatmap
heatmap_data = sentiment_means.melt(id_vars='category', var_name='Sentiment', value_name='Score')
heatmap_data_pivot = heatmap_data.pivot(index='category', columns='Sentiment', values='Score')

# Ensure all categories and sentiments are present
heatmap_data_pivot = heatmap_data_pivot.reindex(columns=['neg_sentiment', 'neu_sentiment', 'pos_sentiment', 'compound_sentiment'])

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data_pivot, annot=False, cmap="YlGnBu", cbar=True, linewidths=.5)
plt.title('Heatmap of Mean Sentiment Scores by Category')
plt.show()
```

### Conclusion

Based on the above charts and data, we can draw some deeper conclusions:

1. **Business Category**:
   Business talks exhibit significant positivity and motivational content. The highest positive and compound sentiment scores reflect that these talks aim to inspire innovation, encourage entrepreneurship, and showcase success stories, thereby evoking positive emotions and actions from the audience.

2. **Science Category**:
   Science talks have higher neutral and negative sentiment scores, indicating that these talks more frequently discuss challenges and issues in the scientific field. They tend to maintain objectivity and rigor, focusing on the complexity and uncertainty of scientific research.

3. **Technology Category**:
   Technology talks have the highest neutral sentiment score, reflecting that these talks tend to be more neutral and objective in conveying information. They avoid emotional bias to ensure the accuracy and comprehensiveness of the information.
